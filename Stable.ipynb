```python
import pandas as pd
import numpy as np
import torch
from huggingface_hub import notebook_login
from diffusers import StableDiffusionPipeline
```

```python
CUDA_LAUNCH_BLOCKING=1
```

```python
class CFG:
    device = "cuda" if torch.cuda.is_available() else "cpu"
    seed = 42
    generator = torch.Generator(device).manual_seed(seed)
    model_id = "stabilityai/stable-diffusion-2"

CFG = CFG()
```

```python
notebook_login()
```

```python
model_pipe = StableDiffusionPipeline.from_pretrained(CFG.model_id, torch_dtype=torch.float16, revision="fp16", use_auth_token=True)
model_pipe = model_pipe.to(CFG.device)
```

```python
def generate_image(prompt, model):
    return model(prompt).images[0]
```

```python
import gc
gc.collect()
```

```python
generate_image("hyper realistic photo of very friendly and dystopian crater", model_pipe)
```
